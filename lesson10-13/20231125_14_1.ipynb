```python
import numpy as np
import pandas as pd
scroes = np.random.randint(50, 100,size=(50,5))
scores_df = pd.DataFrame(scroes,
             columns=['國文', '英文', '數學', '地理', '歷史'],
             index=range(1,51))
scores_df
sum_values = scores_df.sum(axis=1)
mean_values = scores_df.mean(axis=1)
scores_df['總分'] = sum_values
scores_df['平均'] = mean_values
```

```python
scores_df
```

```python
#總分由小到大sort_values  ascemding=false指由大到小 iloc[:5]前五比
scores_df.sort_values(by='總分',ascending=False).iloc[:5]
#scores_df.sort_values(by='總分').iloc[:5]
```

```python
#使用subscript[],過濾資料
scores_df[scores_df['平均'] < 80]
```

```python
#平均80以下,歷史不及格的
#scores_df[{scores_df['平均'] < 80} & {scores_df['歷史'] < 60}]
scores_df[(scores_df['平均'] < 80) & (scores_df['歷史'] < 60)]
```
