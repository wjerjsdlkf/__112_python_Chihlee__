```python
import numpy as np
import pandas as pd
#(50,100) 50到100
#size=(50,5) 50位的隨機5個分數
scroes = np.random.randint(50, 100,size=(50,5))
scores_df = pd.DataFrame(scroes,
             columns=['國文', '英文', '數學', '地理', '歷史'],
             index=range(1,51))#不包含51
scores_df
```

```python
#'國文', '英文', '數學', '地理', '歷史'


scores_df['總分'] = scores_df['國文'] + scores_df['英文'] + scores_df['數學'] + scores_df['地理'] + scores_df['歷史']
scores_df['平均'] = scores_df['總分'] / 5
scores_df
```

```python
scores_df = pd.DataFrame(scroes,
             columns=['國文', '英文', '數學', '地理', '歷史'],
             index=range(1,51))
scores_df
sum_values = scores_df.sum(axis=1)#pandas說明書內axis=1指從列1開始不計算列0
mean_values = scores_df.mean(axis=1)#mean 平均
```

```python
scores_df['總分'] = sum_values
scores_df['平均'] = mean_values
scores_df
```
