```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.arange(12).reshape((4,3)),
             index= [['a','a','b','b'],[1, 2, 1, 2]],
             columns = [['台北','台北','台中'],['Green','Red','Green']])
df
```

```python
df.index.names = ["key1", "key2"]
df.columns.names = ["縣市","顏色"]
df
```

```python
df['台北']['Green']
```

```python
df.loc['a']['台中']
```

```python
#key1跟key2對調
df.swaplevel('縣市','顏色',axis=1)
```

```python
#索引排序
df.sort_index(level='key2')
```

```python
#groupby 索引加總
df.groupby(level='key1').sum()
```

```python
#如果索引是欄名要用".T"去轉成列名再來加總
df.T.groupby(level='顏色').sum()
```
